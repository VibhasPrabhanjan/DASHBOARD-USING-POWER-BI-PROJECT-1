```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

def load_movie_data(filepath="/content/deepseek_csv_20250727_9f4b4d.txt"):
    """Load movie data from CSV file"""
    try:
        return pd.read_csv(filepath)
    except FileNotFoundError:
        print(f"Error: File '{filepath}' not found.")
        return None

def create_dashboard(df):
    """Create the movie ratings dashboard"""
    plt.figure(figsize=(15, 12))
    plt.suptitle('Movie Ratings Dashboard', fontsize=16, y=1.02)

    # Create grid layout
    gs = GridSpec(3, 2, figure=plt.gcf(), height_ratios=[1, 1, 1.5])

    # 1. Top 5 Highest Rated Movies (Table)
    ax1 = plt.subplot(gs[0, 0])
    top5 = df.sort_values('Rating', ascending=False).head(5)[['Title', 'Rating']]
    ax1.axis('off')
    ax1.table(cellText=top5.values,
              colLabels=top5.columns,
              loc='center',
              cellLoc='center',
              colWidths=[0.7, 0.3])
    ax1.set_title('Top 5 Highest Rated Movies', pad=20)

    # 2. Average Rating by Genre (Bar Chart)
    ax2 = plt.subplot(gs[0, 1])
    avg_rating = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
    colors = plt.cm.viridis_r(range(len(avg_rating)))
    avg_rating.plot(kind='bar', ax=ax2, color=colors)
    ax2.set_title('Average Rating by Genre')
    ax2.set_ylabel('Average Rating')
    ax2.set_ylim(0, 10)
    plt.xticks(rotation=45, ha='right')

    # 3. Movie Count by Genre (Pie Chart)
    ax3 = plt.subplot(gs[1, 0])
    genre_counts = df['Genre'].value_counts()
    genre_counts.plot(kind='pie', ax=ax3, autopct='%1.1f%%',
                     startangle=90, colors=plt.cm.Pastel1.colors)
    ax3.set_title('Movie Count by Genre')
    ax3.set_ylabel('')  # Remove y-label

    # 4. Rating Distribution (Histogram)
    ax4 = plt.subplot(gs[1, 1])
    df['Rating'].plot(kind='hist', ax=ax4, bins=10,
                      color='skyblue', edgecolor='black')
    ax4.set_title('Rating Distribution')
    ax4.set_xlabel('Rating')
    ax4.set_ylabel('Number of Movies')
    ax4.set_xlim(0, 10)

    # 5. Reviews vs Rating (Scatter Plot)
    ax5 = plt.subplot(gs[2, :])
    genres = df['Genre'].unique()
    colors = plt.cm.tab10(range(len(genres)))

    for genre, color in zip(genres, colors):
        genre_data = df[df['Genre'] == genre]
        ax5.scatter(genre_data['Reviews'], genre_data['Rating'],
                   color=color, label=genre, alpha=0.7, s=100)

    ax5.set_title('Reviews vs Rating by Genre')
    ax5.set_xlabel('Number of Reviews (millions)')
    ax5.set_ylabel('Rating')
    ax5.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax5.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig('movie_ratings_dashboard.png', dpi=300, bbox_inches='tight')
    print("Dashboard saved as 'movie_ratings_dashboard.png'")
    plt.show()

def main():
    """Main function to run the dashboard"""
    print("Movie Ratings Dashboard Generator")
    print("--------------------------------")

    # Load data
    df = load_movie_data()
    if df is None:
        return

    # Display basic info
    print(f"\nLoaded {len(df)} movies")
    print("\nSample data:")
    print(df.head(3))

    # Create dashboard
    print("\nGenerating dashboard...")
    create_dashboard(df)

if __name__ == "__main__":
    main()
```
